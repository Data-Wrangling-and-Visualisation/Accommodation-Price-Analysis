```python
import pandas as pd
from geopy.geocoders import Nominatim
from time import sleep

# Load CSV
file_path = "cian_sochi_all1-57.csv"  # Change to your actual file path
df = pd.read_csv(file_path)

# Initialize Geocoder
geolocator = Nominatim(user_agent="geo_finder")

# Function to get coordinates
def get_coordinates(street, house_number):
    try:
        address = f"{street} {house_number}, Сочи, Россия"
        location = geolocator.geocode(address, timeout=10)
        if location:
            return pd.Series([location.latitude, location.longitude])
    except Exception as e:
        print(f"Error for {address}: {e}")
    return pd.Series([None, None])  # If not found, return None

# Apply function to each row
df[["latitude", "longitude"]] = df.apply(lambda row: get_coordinates(row["street"], row["house_number"]), axis=1)

# Save updated file
output_file = "cian_sochi_all1-57_lat.csv"
df.to_csv(output_file, index=False)

print(f"Updated file saved as {output_file}")
```
