```python
import pandas as pd


CBR_df = pd.read_csv('../data/RUS_CBR_mortgage_data_Krasnodarskii_krai.tsv', delimiter='\t')
```

```python
columns_RU = ["Отчетный период"
              "Ставки (% годовых)",
              "Количество кредитов",
              "Объем кредитов (млн руб.)",
              "Задолженность по кредитам (млн руб.)",
              "Просроченная задолженность по кредитам (млн руб.)",
              "Средневзвешенный срок по кредитам (месяцев)"]
columns_EN = ["Period",
              "Yearly rate (%)",
              "New mortgages",
              "New mortgage amount (millions)",
              "Total debt (millions)",
              "Overdue debt (millions)",
              "Mean new mortgage term (months)"]
columns_EN_export = ["Month",
                     "Year",
                     "Yearly rate (%)",
                     "New mortgages",
                     "New mortgage amount (millions)",
                     "Total debt (millions)",
                     "Overdue debt (millions)",
                     "Mean new mortgage term (months)"]
```

```python
CBR_df.columns = columns_EN
```

```python
CBR_df[["Month", "Year"]] = CBR_df["Period"].str.split(expand=True)
rus_month_to_num = {
        "Январь": 1,
        "Февраль": 2,
        "Март": 3,
        "Апрель": 4,
        "Май": 5,
        "Июнь": 6,
        "Июль": 7,
        "Август": 8,
        "Сентябрь": 9,
        "Октябрь": 10,
        "Ноябрь": 11,
        "Декабрь": 12
    }
CBR_df["Month"] = CBR_df["Month"].apply(lambda x: rus_month_to_num[x])
```
